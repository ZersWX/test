```python
#画出全球各行业的温室气体排放图的柱状图
import pandas as pd
import matplotlib.pyplot as pl
import pandas as pd
data=pd.read_csv("/home/mw/input/greenhousegas6025/Greenhouse_Gas.csv")
column_world=['Country/Region','Industry', 'F2010', 'F2011', 'F2012', 'F2013', 'F2014',
       'F2015', 'F2016', 'F2017', 'F2018', 'F2019', 'F2020', 'F2021']
data_process=data[column_world]
temp=[]
v=data_process.values
temp=[]
for item in v:
    if item[0]=="World":
        temp.append(item)
data_world=pd.DataFrame(columns=column_world,data=temp)

#print(data_world)
#计算每一个行业每年温室气体排放总量
n=int(len(data_world)/5)
temp1=[]
industrys=[]
for i in range(n):
    start=i*5
    end=(i+1)*5
    industry_name=data_world.iloc[start:start+1,1]
    industrys.append(industry_name.values)
    #print(data_world.iloc[i*5:(i+1)*5,2:])
    #求总排放量
    #print(industry_name)
    sum_number=data_world.iloc[start:end,2:].sum()
    #print(sum_number)
    temp1.append(sum_number)

final_data=pd.DataFrame(columns=['F2010', 'F2011', 'F2012', 'F2013', 'F2014',
       'F2015', 'F2016', 'F2017', 'F2018', 'F2019', 'F2020', 'F2021'],data=temp1)
#print(final_data)
print(final_data)
final_data.to_csv("./data.csv")
bar_temp=final_data.plot.bar(rot=0, figsize=(10, 6), width=0.9)
# 自定义横轴标签及其位置
bar_temp.set_xticks(range(len(final_data.index)))
for x in industrys:
    print(x[0])
bar_temp.set_xticklabels(industrys)

# 设置 x 轴标签的角度、字体大小和距离
pl.xticks(rotation=45, fontsize=12, ha='right')
pl.title("2010到2021年世界各行业温室气体的排放总量变化")
pl.show()

```
