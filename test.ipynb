```python
import numpy as np

def is_all_nan(arr):
    """判断一个数组是否全为NaN值"""
    nan_count = np.sum(np.isnan(arr))
    if nan_count!=0 and nan_count <= arr.size*0.2:
        return True
    else:
        return False
```

```python
#饼图各能源转型方式的使用频率

import pandas as pd
import matplotlib.pyplot as plt

# 从CSV文件中读取数据
data = pd.read_csv('/home/mw/project/data0_2000-2020.csv')

# 获取需要绘制饼图的列数据
column_data = data['Technology']

# 计算每个数据出现的次数
counts = column_data.value_counts()

# 定义不同类型能源转型方式对应的颜色
colors = ['#FFA07A', '#ADD8E6', '#98FB98', '#FFE4E1', '#CD5C5C', '#90EE90',
          '#8A2BE2', '#00CED1', '#F08080', '#FFDAB9']
# 绘制饼图
fig, ax = plt.subplots(figsize=(8, 6))
explode = [0.1 if i == 0 else 0.05 for i in range(len(counts))]  # 设置分离半径
wedgeprops = {'width': 0.7, 'edgecolor': 'w', 'linewidth': 2}  # 设置楔形块属性
ax.pie(counts.values, labels=counts.index.tolist(), autopct='%1.1f%%', startangle=90,
       colors=colors, explode=explode, wedgeprops=wedgeprops, shadow=True)

# 添加图表标题
plt.title('Pie Chart of Energy Transition Modes', fontsize=16)
plt.show()
```

```python
# 读取CSV文件
df = pd.read_csv('/home/mw/input/temp6990/Temp.csv')

# 选取需要的列
df = df[['Country/Region'] + [f'F{i}' for i in range(1961, 2022)]]

# 将 'Country/Region' 列设为索引
df = df.set_index('Country/Region')

# 转换数据格式
df = df.T

# 去除空值超过10%的列
df = df.dropna(thresh=len(df)*0.9, axis=1)

# 使用插值法填充空值
df = df.interpolate(method='linear', limit_direction='both')

# 计算每个国家的温度变化范围
diffs = df.max(axis=0) - df.min(axis=0)

# 找出温度变化范围最大的前5个国家
top_countries = diffs.sort_values(ascending=False).head().index.tolist()

# 为这些国家绘制时间序列图
plt.figure(figsize=(15,8))
for country in top_countries:
    plt.plot(df[country], label=country)

plt.title('Temperature Changes of Top 5 Countries')
plt.xlabel('Year')
plt.ylabel('Temperature (Celsius)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import statsmodels.api as sm

# 读取数据
data = pd.read_csv('/home/mw/input/temp6990/Temp.csv')

# 选取数值型数据列，并处理缺失值
num_data = data.select_dtypes(include='number')
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputed_data = imputer.fit_transform(num_data.iloc[:, 1:])

# 标准化处理
scaler = StandardScaler()
scaled_data = scaler.fit_transform(imputed_data)

# PCA降维
n_components = 2
pca = PCA(n_components=n_components)
pca_data = pca.fit_transform(scaled_data)

# 计算轮廓系数选择K值
k_list = list(range(2, 10))
silhouette_scores = []
for k in k_list:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pca_data)
    score = silhouette_score(pca_data, kmeans.labels_)
    silhouette_scores.append(score)
best_k = k_list[np.argmax(silhouette_scores)]

# 聚类
kmeans = KMeans(n_clusters=best_k)
kmeans.fit(pca_data)

# 可视化聚类结果
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k'] * 2
for i in range(len(kmeans.labels_)):
    plt.scatter(pca_data[i, 0], pca_data[i, 1], color=colors[kmeans.labels_[i]])
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title('Clustering Result')
plt.show()

# 找到温度变化最大的几个国家
cluster_centers = scaler.inverse_transform(pca.inverse_transform(kmeans.cluster_centers_))
temp_diffs = np.abs(cluster_centers[:, -1] - cluster_centers[:, 1])
sorted_indices = np.argsort(temp_diffs)[::-1]
top_n = 3
top_n_indices = sorted_indices[:top_n]

# 对选定的几个国家进行时间序列预测分析
print(top_n_indices)
country_names = data.iloc[top_n_indices, 1].tolist()
print(country_names)


for i in top_n_indices:
    # 处理缺失值
    imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
    imputed_row = imputer.fit_transform(num_data.iloc[i:i+1, 1:])

    # 时间序列分析
    ts_data = pd.Series(imputed_row[0])
    ts_data[ts_data.isna()] = ts_data.interpolate()
    ts_train = ts_data[:57]

    # 选择ARIMA模型阶数
    p = 1
    d = 1
    q = 1

    # 模型拟合和预测未来5年
    arima_model = sm.tsa.ARIMA(ts_train, order=(p, d, q))
    arima_result = arima_model.fit(disp=-1)
    n_preds = 5 * 12
    preds = arima_result.predict(57, 56+n_preds, typ='levels', dynamic=True)

    # 可视化结果
    plt.figure(figsize=(8, 6))
    plt.plot(ts_train.index, ts_train.values, label='train')
    plt.plot(preds.index, preds.values, label='prediction')
    plt.legend()
    plt.xlabel('Year')
    plt.ylabel('Temperature')
    plt.title(f'Prediction for Country/Region {data.iloc[i, 0]}')
    plt.show()

    # 模型评估
    print(f'Model evaluation for Country/Region {data.iloc[i, 0]}:')
    print(f'ARIMA({p},{d},{q})')
    print(f'MSE: {((preds - ts_data[57:])**2).mean():.2f}\n')
```
